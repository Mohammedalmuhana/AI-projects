```python
def tokenizer(col):
    word_id={}
    id_word = {}
    for id,word in enumerate(col.unique()):
        word_id[word]=id 
        id_word[id] = word
    return word_id,id_word
```

```python
# Get the label to id and id to label dictdictionary 
Sex_id,id_Sex = tokenizer(df.Sex)
# Convert the labels to ids using the label_id dictionary 
df.Sex.replace(Sex_id,inplace=True)

# Get the label to id and id to label dictdictionary 
Ethnicity_id,id_Ethnicity = tokenizer(df.Ethnicity)
# Convert the labels to ids using the label_id dictionary 
df.Ethnicity.replace(Ethnicity_id,inplace=True)
```
